```python
#Need to see how many ratings the typical user has before

import pandas as pd
import csv


#make sure that movies are only rated once per user...
user_to_list_of_movie_and_rating = dict()

for i in range(4):
    file = open("extra_data/combined_data_"+str(i+1)+".txt", 'r')
    lines = file.readlines()
    file.close()





    current_movie_id = -1

    for item in lines:
        if item[-2] == ":":
            current_movie_id = item[:-2]
        else: 
            values = item.split(",")
            if(values[0] in user_to_list_of_movie_and_rating.keys()):
                user_to_list_of_movie_and_rating[values[0]].append([current_movie_id, values[1]])
            else: 
                user_to_list_of_movie_and_rating[values[0]] = [[current_movie_id, values[1]]]




print(user_to_list_of_movie_and_rating['1488844'])
print(user_to_list_of_movie_and_rating['822109'])


dup = False

#need to check for duplicates here:
for item in user_to_list_of_movie_and_rating.keys(): 
    user_to_list_of_movie_and_rating[item] = sorted(user_to_list_of_movie_and_rating[item])
    last_item = -1
    for i in range(user_to_list_of_movie_and_rating[item]):
        if(user_to_list_of_movie_and_rating[item][i][0] == last_item):
            last_item = user_to_list_of_movie_and_rating[item][i][0]
            del user_to_list_of_movie_and_rating[item][i]
            i-=1
            dup = True
        else: 
            last_item = user_to_list_of_movie_and_rating[item][i][0]


if dup:
    print("dup")


sum = 0
cnt = 0
for item in user_to_list_of_movie_and_rating.keys():
    sum+= len(user_to_list_of_movie_and_rating[item])
    cnt+=1

print(sum/cnt)

```
